```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
```

```python
page = requests.get(url)
```

```python
page
```

```python
# print(page.text)
```

```python
soup = BeautifulSoup(page.text,'html')
#print(soup)
```

```python
# print(soup.prettify())
```

```python
table1 = soup.find_all('table')[1]
table1
```

```python
table1.find_all('th')
```

```python
# , class_ ='wikitable sortable jquery-tablesorter'
```

```python
columns = table1.find_all('th')
columns
```

```python
columns_title = [title.text.strip() for title in columns]
```

```python
df = pd.DataFrame(columns = columns_title)
df
```

```python
columns_data = table1.find_all("tr")
#columns_data
```

```python
#columns_data[1:]
```

```python
for row in columns_data[1:]:
    row_data = row.find_all("td")
    ind_row_data = [i.text.strip() for i in row_data]
    length = len(df)
    df.loc[length] = ind_row_data
```

```python
df
```

```python
df.to_csv("Table_1.csv", index=False)
```
