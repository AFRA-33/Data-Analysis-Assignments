## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
df.head(10)
```

```python
df['income'] = np.where(df['income'] == '<=50K', 0, 1)

# Check the updated DataFrame
print(df['income'].head())
```

## ***18. Which Workclass getting The Highest Salary***

```python
high_salary_of_wrokclass = df.groupby('workclass')['income'].mean().idxmax()
print(f"The workclass with the highest average salary is: {high_salary_of_wrokclass}")
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
df['income'].value_counts()
```

```python
high_salary_of_gender = df.groupby('gender')['income'].mean().idxmax()
print(f"The workclass with the highest average salary is: {high_salary_of_gender}")
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='gender',hue='income',data=df)
plt.title("Countplot of Income vs Gender")
plt.xlabel("Gender")
plt.ylabel("Income")
plt.xticks(rotation = 90)
plt.show()
```
